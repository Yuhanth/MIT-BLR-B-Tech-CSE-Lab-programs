```python
import pandas as pd

df = pd.read_csv('cereals.csv')

numeric_cols = ['protein', 'fat', 'sodium', 'fiber', 'carbohydrates', 'sugars', 'vitamins']
df[numeric_cols] = df[numeric_cols].astype(float)

print("5-Number Summary (Original):")
summary_orig = df[numeric_cols].describe().loc[['min', '25%', '50%', '75%', 'max']]
print(summary_orig)

df_missing = df.copy()
for col in numeric_cols:
    actual_mean = df_missing.loc[df_missing[col] != -1, col].mean()
    df_missing[col] = df_missing[col].replace(-1, actual_mean)

print("\n5-Number Summary (After Missing Value Treatment):")
summary_missing = df_missing[numeric_cols].describe().loc[['min', '25%', '50%', '75%', 'max']]
print(summary_missing)

df_noisy = df_missing.copy()
for col in numeric_cols:
    Q1 = df_noisy[col].quantile(0.25)
    Q3 = df_noisy[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    col_median = df_noisy[col].median()
    df_noisy.loc[(df_noisy[col] < lower_bound) | (df_noisy[col] > upper_bound), col] = col_median

print("\n5-Number Summary (After Noisy Value Treatment):")
summary_noisy = df_noisy[numeric_cols].describe().loc[['min', '25%', '50%', '75%', 'max']]
print(summary_noisy)
```
