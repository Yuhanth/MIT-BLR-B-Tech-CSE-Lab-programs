```python
numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))

print("Largest number:", max(numbers))
print("Smallest number:", min(numbers))
print("Sum of numbers:", sum(numbers))
print("Average of numbers:", sum(numbers) / len(numbers))

target = int(input("Enter the number to count: "))
print(f"The number {target} occurs {numbers.count(target)} times in the list.")
```

```python
my_tuple = (10, 20, 30, 40, 50)
print("Original tuple:", my_tuple)

try:
    my_tuple[2] = 99
except TypeError as e:
    print("Error:", e)
    print("Explanation: Tuples are immutable in Python, meaning their elements cannot be changed after creation.")
```

```python
world_cup_winners = {
    2011: "India",
    2015: "Australia",
    2019: "England",
    2023: "Australia"
}

win_count = {}
for country in world_cup_winners.values():
    win_count[country] = win_count.get(country, 0) + 1

best_country = max(win_count, key=win_count.get)

print("Best performing country:", best_country)
print("Unique list of countries that have won the World Cup:", list(set(world_cup_winners.values())))
```

```python
sentence = input("Enter a sentence: ")
words = sentence.split()
word_freq = {}

for word in words:
    word = word.lower()
    word_freq[word] = word_freq.get(word, 0) + 1

print("Word frequencies:")
print(word_freq)
```

```python
set1 = set(map(int, input("Enter numbers for Set 1 (separated by spaces): ").split()))
set2 = set(map(int, input("Enter numbers for Set 2 (separated by spaces): ").split()))

print("Set 1:", set1)
print("Set 2:", set2)
print("Union:", set1.union(set2))
print("Intersection:", set1.intersection(set2))
print("Difference (Set1 - Set2):", set1.difference(set2))
```
