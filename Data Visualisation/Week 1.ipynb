```python
def isPrime(a):
    prime = True
    for i in range(2,int(a/2)+1):
        if a%i==0:
            prime = False
            break
    if prime:
        print(f"{a} is a prime number")
    else:
        print(f"{a} is not a prime number")
a = int(input("Enter a number: "))
isPrime(a)
```

```python
def seqSum():
    seq=[]; sum = 0
    for i in range(1,100+1):
        seq.append(i)
    for i in seq:
        if i%2==0:
            sum+=i
        else:
            print(i, end=" ")
    print(f"\nSum of even numbers: {sum}")
seqSum()
```

```python
def countVowels(string):
    vow=cons=0
    for i in string:
        if i.isalpha():
            if i.lower() in ['a','e','i','o','u']:
                vow+=1
            else:
                cons+=1
    print(f"No. of vowels = {vow}\nNo. of consonants = {cons}")
countVowels(input("Enter a sring :"))
```
